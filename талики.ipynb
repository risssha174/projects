```python
import numpy as np
import matplotlib.pyplot as plt

#температуры брались за 2016г.

cities = {
    "Якутск": [-34.8,-35,-14.8,-0.1,7.6,16,18.1,13.4,6.9,-7.4,-26.5,-36.9],
    "Оймякон": [-43,-44.7,-27.5,-10.1,4.8,13.4,15.2,10.2,1.9,-10.8,-34.3,-44.2],
    "Салехард": [-23.7,-10.3,-12.8,-2.1,1.2,14.1,19.4,14.6,9.3,-0.2,-17.7,-26.2],
    "Ханты-Мансийск": [-20,-7.9,-7.4,4.3,9.7,17.8,20,18.6,10.7,-0.8,-18.2,-23.6],
    "Диксон": [-17.6,-16.2,-17.3,-9.2,-5.4,3.7,9.9,8,7.2,-0.8,-18.6,-30.6]
}

k = {
    "Якутск": 0.8,
    "Оймякон": 0.6,
    "Салехард": 1.0,
    "Ханты-Мансийск": 1.2,
    "Диксон": 0.5
}

months = np.arange(1, 13)

for city, temps in cities.items():

    cumulative = []
    total = 0

    for T in temps:
        if T > 0:
            total += T * 30
        cumulative.append(total)

    depths = 0.05 * k[city] * np.sqrt(cumulative)

    plt.figure()
    plt.plot(months, depths, marker='o')

    plt.gca().invert_yaxis()  # глубина идет вниз по графику

    plt.title(city)
    plt.xlabel("Месяц")
    plt.ylabel("Глубина (м)")
    plt.grid()

    plt.show()

    print(city, "max depth:", round(max(depths), 2), "м")
```
